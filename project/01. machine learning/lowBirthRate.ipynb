```python
import pandas as pd

# 데이터 준비
education = pd.read_csv('data/사교육비.csv')
time = pd.read_csv('data/근로시간.csv')
house = pd.read_csv('data/주거비.csv')
education.head()
```

```python
import numpy as np

merged_df = pd.merge(education, time, on=['시점', '지역별'], how='outer')
merged_df = pd.merge(merged_df, house, on=['시점', '지역별'], how='outer')
merged_df['총소득'] = merged_df['총소득'].replace(0, np.nan)
merged_df.head()
```
