```python
import pandas as pd

# 예시 데이터프레임 생성
data = {'Category': ['A', 'B', 'C', 'D'],
        'Values': [10, 20, 15, 25]}
df = pd.DataFrame(data)
df
```

### matplotlib

```python
import matplotlib.pyplot as plt

# 막대 그래프
plt.figure(figsize=(8, 6))
plt.bar(df['Category'], df['Values'], color='skyblue')
plt.title('Matplotlib Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

### seaborn

```python
import seaborn as sns

# 막대 그래프
plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Values', data=df, palette='Blues')
plt.title('Seaborn Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

### plotly

```python
import plotly.express as px

# 막대 그래프
fig = px.bar(df, x='Category', y='Values', title='Plotly Bar Chart')
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# 예시 데이터프레임 생성
data = {
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'D', 'D', 'D'],
    'Subcategory': ['X', 'Y', 'Z', 'X', 'Y', 'Z', 'X', 'Y', 'Z', 'X', 'Y', 'Z'],
    'Values': [10, 15, 13, 22, 24, 23, 17, 19, 15, 26, 27, 25],
    'Region': ['North', 'North', 'North', 'East', 'East', 'East', 'South', 'South', 'South', 'West', 'West', 'West']
}
df = pd.DataFrame(data)

# Plotly로 복잡한 차트 생성
fig = px.bar(
    df,
    x='Subcategory',
    y='Values',
    color='Region',
    facet_col='Category',
    title='Complex Plotly Bar Chart with Facet and Hover Info',
    hover_data=['Category', 'Region'],  # Hover시 표시할 추가 정보
    barmode='group'  # 그룹별로 막대가 나란히 출력되도록 설정
)

# 레이아웃 조정
fig.update_layout(
    height=600,
    width=1000,
    showlegend=True,  # 범례 표시
    title_font_size=20,
)

# 차트 출력
fig.show()
```
