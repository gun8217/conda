### fundamental

`BPS/PER/PBR/EPS/DIV/DPS
- BPS[주당순자산가치 (Book-value per share)]:기업의 순자산을 발행 주식수로 나눈 것
- PER : 주가 / 주당 순이익 (버는 것에 비해 가격이 어떤가?)
- PBR : 주가 / BPS -> 주가 순자산 비율 -> 자산에 비해 주가가 얼마나 비싼가?
  - PBR > 1 : 고평가
  - PBR < 1 : 저평가
- EPS :1주당 이익
- DIV 배당수익률 (Divedend yield ratio) : 주당 배당금 / 현재주가
- DPS (Divedend per share) : 1 주당 배당금

### 실습 (배당 종목 선정하기)

1. 시가총액 1조 이상(특정 규모 이상이 되는가?)

```python
market_cap = stock.get_market_cap('20241118')
df1 = market_cap[market_cap['시가총액'] >= 1000000000000]
df1.head()
```

2. 돈을 잘 버는 능력이 있는가? (ROE)

```python
fundamental = stock.get_market_fundamental('20241118')
fundamental['ROE'] = fundamental['EPS'] / fundamental['BPS'] * 100
fundamental.head()
```

```python
df2 = fundamental[(fundamental['ROE'] >= 10) & fundamental.index.isin(df1.index)]
len(df2)
```

3. 현재 배당률 3% 이상

```python
df3 = df2[df2['DIV'] >= 3]
len(df3)
```

4. 최근 5년간 배당을 지급했는지?

```python
dates = ['20191230', '20201230', '20211230','20221229','20231228']

import pandas as pd
result = pd.DataFrame()
for date in dates:
  df = stock.get_market_fundamental(date)[['DPS']].rename(columns = {'DPS':date})
  df = df[df[date] > 0] # 배당을 지급한 데이터만 남깁니다.
  result = pd.concat([result, df], axis = 1)
result.head()
```

```python
df4 = result[result.index.isin(df3.index)].dropna(axis = 0)
df4
```
