### resample()
- upsampling : ex) 월간 -> 일
- downsampling : ex) 주간 -> 월 or 일간 -> 주간

```python
import FinanceDataReader as fdr
df = fdr.DataReader('005380', '2019')
df.head()
```

```python
df.resample(rule = 'Y').mean()
```

```python
df.resample(rule = 'Q').mean()
```

```python
df.resample(rule = 'Q').max()
```

```python
resample_df = df.resample(rule = 'Y').mean()
resample_df.head()
```

```python
import plotly.express as px

fig = px.bar(resample_df, x=resample_df.index, y='Close')
fig.show()
```

### shift

```python
AAPL_close = df[['Close']].copy()
AAPL_close.loc[:, 'Close_p1'] = AAPL_close.loc[:,'Close'].shift(1)
AAPL_close.loc[:, 'Close_m1'] = AAPL_close.loc[:,'Close'].shift(-1)
AAPL_close.head()
```

### asfreq

```python
df.asfreq('30S')
```

```python
df2 = df.loc['2023-01-02':, "Close"]
df2.head()
```

```python
ffill = df2.asfreq('3h').ffill()
bfill = df2.asfreq('3h').bfill()
interpolate = df2.asfreq('3h').interpolate()
```

```python
import plotly.graph_objects as go

# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=ffill.index, y=ffill.values,
                    mode='lines',
                    name='ffill'))
fig.add_trace(go.Scatter(x=bfill.index, y=bfill.values,
                    mode='lines',
                    name='bfill'))
fig.add_trace(go.Scatter(x=interpolate.index, y=interpolate.values,
                    mode='lines', name='interpolate'))

fig.show()
```

### rolling

```python
MA5 = df['Close'].rolling(window = 5).mean()
MA30 = df['Close'].rolling(window = 30).mean()
MA5.head(10)
```

```python
# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=MA5.index, y=MA5.values,
                    mode='lines',
                    name='MA5'))
fig.add_trace(go.Scatter(x=MA30.index, y=MA30.values,
                    mode='lines',
                    name='MA30'))
fig.add_trace(go.Scatter(x=df.index, y=df['Close'],
                    mode='lines', name='Close'))

fig.show()
```

### Expanding(누적평균)

```python
EX30 = df['Close'].expanding(30).mean()
```

```python
# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=EX30.index, y=EX30.values,
                    mode='lines',
                    name='EX30'))
fig.add_trace(go.Scatter(x=MA30.index, y=MA30.values,
                    mode='lines',
                    name='MA30'))
fig.add_trace(go.Scatter(x=df.index, y=df['Close'],
                    mode='lines', name='Close'))

fig.show()
```

### 실습 (2021년 이후 데이터)



1. 테슬라(TSLA)의 종가 전일 대비 변화율을 히스토그램으로 시각화

```python
TSLA = fdr.DataReader('TSLA', '2021-01-01')
TSLA['Changes'] = TSLA['Close'].pct_change()
TSLA.head()
```

```python
import plotly.express as px
fig = px.histogram(TSLA, x="Changes")
fig.show()
```

2. 테슬라(TSLA)의 이동평균 5, 20, 60일을 종가와 함께 라인차트로 시각화

```python
TSLA['MA5'] = TSLA['Close'].rolling(window = 5).mean()
TSLA['MA20'] = TSLA['Close'].rolling(window = 20).mean()
TSLA['MA60'] = TSLA['Close'].rolling(window = 60).mean()
TSLA.head()
```

```python
# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=TSLA.index, y=TSLA['MA5'],
                    mode='lines',
                    name='MA5'))
fig.add_trace(go.Scatter(x=TSLA.index, y=TSLA['MA20'],
                    mode='lines',
                    name='MA20'))
fig.add_trace(go.Scatter(x=TSLA.index, y=TSLA['MA60'],
                    mode='lines', name='MA60'))
fig.add_trace(go.Scatter(x=TSLA.index, y=TSLA['Close'],
                    mode='lines', name='Close'))
fig.show()
```

3. 테슬라(TSLA)의 종가를 분기단위로 리셈플링 한 뒤에 막대차트로 시각화

```python
resample_QE = TSLA['Close'].resample('QE').mean()
resample_QE.head()
```

```python
import plotly.express as px

fig = px.bar(resample_QE, x=resample_QE.index, y=resample_QE.values)
fig.show()
```
