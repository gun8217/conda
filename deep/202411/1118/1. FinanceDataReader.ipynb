### 실습
1. 최근 2년(2022년 이후) 네이버 주가(Close)를 히스토그램으로 시각화
2. 최근 2년(2022년 이후) 네이버와 카카오의 주가(Close)를 스케터 차트로 시각화
3. 최근 2년(2022년 이후) 삼성전자, 현대차, 네이버 주가(Close)를 가장 첫날을 100으로 지수화 한 뒤에 라인차트로 시각화

```python
# 1
naver = fdr.DataReader("035420", "2022")
naver.head()
```

```python
import plotly.express as px
fig = px.histogram(naver, x="Close")
fig.show()
```

```python
# 2. 최근 2년(2022년 이후) 네이버와 카카오의 주가(Close)를 스케터 차트로 시각화
naver = fdr.DataReader("035420", "2022")
kakao = fdr.DataReader("035720", "2022")

naver.head(), kakao.head()
```

```python
fig = px.scatter(x=naver['Close'], y=kakao['Close'])
fig.show()
```

```python
# 3. 최근 2년(2022년 이후) 삼성전자, 현대차, 네이버 주가(Close)를 가장 첫날을 100으로 지수화 한 뒤에 라인차트로 시각화
samsung = fdr.DataReader("005930", "2022")
hy = fdr.DataReader("005380", "2022")
naver = fdr.DataReader("035420", "2022")

samsung.head(), hy.head(), naver.head()
```

```python
import plotly.graph_objects as go

# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=samsung.index, y=samsung['Close'],
                    mode='lines',
                    name='삼성전자'))
fig.add_trace(go.Scatter(x=hy.index, y=hy['Close'],
                    mode='lines', name='현대차'))
fig.add_trace(go.Scatter(x=naver.index, y=naver['Close'],
                    mode='lines',
                    name='네이버'))

fig.show()
```

```python
samsung_100 = (samsung[['Close']] / samsung[['Close']].iloc[0]) * 100
hy_100 = (hy[['Close']] / hy[['Close']].iloc[0]) * 100
naver_100 = (naver[['Close']] / naver[['Close']].iloc[0]) * 100

samsung_100.head(), hy_100.head(), naver_100.head()
```

```python
# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=samsung_100.index, y=samsung_100['Close'],
                    mode='lines',
                    name='삼성전자'))
fig.add_trace(go.Scatter(x=hy_100.index, y=hy_100['Close'],
                    mode='lines', name='현대차'))
fig.add_trace(go.Scatter(x=naver_100.index, y=naver_100['Close'],
                    mode='lines',
                    name='네이버'))

fig.show()
```
