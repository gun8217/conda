1. 2019년 부터 2023년 까지 당기순이익,매출액,	부채총계,	영업이익,	자본총계,	자산총계를 하나의 데이터프레임으로 가져옵니다.

```python
result_df = pd.DataFrame()
for year in range(2019, 2024):
  data = dart.finstate('CJ제일제당', year)
  account_nm_list = ['자산총계', '부채총계', '자본총계', '매출액', '영업이익', '당기순이익']
  data = data[(data['account_nm'].isin(account_nm_list)) & (data['fs_nm'] == '연결재무제표')]

  subset_data = data[['stock_code','bsns_year' ,'account_nm', 'thstrm_amount']]
  pivot = pd.pivot_table(data = subset_data, index = 'bsns_year', columns = 'account_nm', values = 'thstrm_amount', aggfunc = 'first')
  result_int = pivot.apply(lambda x: x.str.replace(',','').astype(int))
  result_df = pd.concat([result_df, result_int])
```

```python
result_df.head()
```

2. 위의 데이터 프레임을 활용하여 부채비율, 당기순이익증가율, ROA, ROE 계산하기
- 자산 = 부채 + 자본

```python
result_df['부채비율'] = (result_df['부채총계'] / result_df['자본총계']) * 100
result_df['당기순이익증가율'] = (result_df['당기순이익'].diff() / result_df['당기순이익'].shift(1)) * 100
result_df['ROA'] = (result_df['당기순이익'] / result_df['자산총계']) * 100
result_df['ROE'] = (result_df['당기순이익'] / result_df['자본총계']) * 100
result_df.head()
```
