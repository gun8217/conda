```python
import FinanceDataReader as fdr
import pandas as pd

df_krx = fdr.StockListing('KRX')
df_krx.head()
```

```python
kospi_top10 = df_krx[(df_krx['Name'] != '삼성전자우') & (df_krx['Market'] == 'KOSPI')].head(10)
kospi_top10
```

```python
kospi_top10_df = pd.DataFrame()
for idx, row in kospi_top10.iterrows():
  data = fdr.DataReader(str(row['Code']), '2022')[['Close']].rename(columns = {'Close' : row['Name']})
  kospi_top10_df = pd.concat([kospi_top10_df,data], axis = 1)
```

```python
kospi_top10_df.dropna(inplace = True)
kospi_top10_df.head()
```

```python
norm = (kospi_top10_df / kospi_top10_df.iloc[0]) * 100
norm['동일비중'] = norm.sum(axis = 1) / len(norm.columns)
norm.head()
```

```python
marcap_ratio = kospi_top10['Marcap'] / kospi_top10['Marcap'].sum()
marcap_ratio.index = norm.iloc[:, :-1].columns
```

```python
norm['시가총액비중'] = norm.iloc[:, :-1].mul(marcap_ratio, axis = 1).sum(axis = 1)
```

```python
import plotly.graph_objects as go

# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=norm.index, y=norm['동일비중'],
                    mode='lines',
                    name='동일비중'))
fig.add_trace(go.Scatter(x=norm.index, y=norm['시가총액비중'],
                    mode='lines',
                    name='시가총액비중'))

fig.show()
```
