# HumanMessage, AIMessage, SystemMessage
- HumanMessage : 사람이 작성한 메세지
- AIMessage : LLM이 대답한 메세지
- SystemMessage : LLM이 가지고 있는 설정(세뇌하는 느낌)

```python
from langchain_upstage import ChatUpstage
from langchain.schema import HumanMessage, AIMessage, SystemMessage

llm = ChatUpstage( model="solar-pro", api_key = '<API_KEY>')

result = llm(
  [
    SystemMessage(content = '당신의 이름은 짱짱맨이며 반말로 대답합니다.'),
    HumanMessage(content='안녕하세요!'),
    AIMessage(content = '안녕?'),
    HumanMessage(content='이름은 뭐야?'),
  ]
)
print(result.content)
```

# PromptTemplate
- 프롬프트를 템플릿 화

```python
from langchain import PromptTemplate
prompt = PromptTemplate(
  template = "{coffee_name} 주세요."
)

print(prompt.format(coffee_name='아이스 아메리카노'))
print(prompt.format(coffee_name='카푸치노'))
```

```python
import time  #← 실행 시간을 측정하기 위해 time 모듈 가져오기
import langchain
from langchain.cache import InMemoryCache  #← InMemoryCache 가져오기
from langchain_upstage import ChatUpstage
from langchain.schema import HumanMessage

langchain.llm_cache = InMemoryCache() #← llm_cache에 InMemoryCache 설정

llm = ChatUpstage( model="solar-pro", api_key = '<API_KEY>')
start = time.time() #← 실행 시작 시간 기록
result = llm([ #← 첫 번째 실행을 수행
    HumanMessage(content="안녕하세요!")
])

end = time.time() #← 실행 종료 시간 기록
print(result.content)
print(f"실행 시간: {end - start}초")

start = time.time() #← 실행 시작 시간 기록
result = llm([ #← 같은 내용으로 두 번째 실행을 함으로써 캐시가 활용되어 즉시 실행 완료됨
    HumanMessage(content="안녕하세요!")
])

end = time.time() #← 실행 종료 시간 기록
print(result.content)
print(f"실행 시간: {end - start}초")
```
