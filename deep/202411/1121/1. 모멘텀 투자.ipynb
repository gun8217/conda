### 코스피 2023년에 수익률이 좋았던 종목 50개를 2024년에 매수하고 현재까지 보유했을 때 수익률

```python
from pykrx import stock
import pandas as pd

market_price_change1 = stock.get_market_price_change("20240202", "20240501", "ALL").sort_values(by = '등락률', ascending = False) # 옵션 1 : 시장 변경 ALL, KOSDAQ, KOSPI
market_price_change1.head()
```

```python
buy_list = market_price_change1.head(5) # 옵션 2 : 종목수 5~100
```

```python
market_price_change2 = stock.get_market_price_change("20240501", "20240508", "ALL").sort_values(by = '등락률', ascending = False) # 옵션 1과 통일
sell_list = market_price_change2.loc[buy_list.index, :]
```

```python
print(market_price_change2['등락률'].mean())
print(sell_list['등락률'].mean())
```

```python
c1 = buy_list.rename(columns = {'등락률' : '보유_등락률'})
c2 = sell_list.rename(columns = {'등락률' : '매도_등락률'})

concat_df = pd.concat([c1, c2], axis = 1)[['보유_등락률', '매도_등락률']]
concat_df.head()
```

```python
import plotly.express as px
fig = px.scatter(x=concat_df['보유_등락률'], y=concat_df['매도_등락률'],
                 labels={'x':'보유_등락률', 'y':'매도_등락률'})
fig.show()
```

- 한국 시장 기준으로 보유 기간은 짧은게 좋을 수 있다.
- 보유 등락률이 너무 높은 경우에는 오히려 매도 등락률이 하락하는 경우가 발생했다.
- 한국 시장 기준으로 탐색 기간도 짧은게 좋을 수 있다.
- 종목수도 줄이는게 좋을 수 있음
